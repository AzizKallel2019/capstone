```python
from confluent_kafka import Consumer, KafkaError, KafkaException
import json

# Kafka Consumer configuration
consumer_config = {
    'bootstrap.servers': 'localhost:9092',
    'group.id': 'my_consumer_group',
    'auto.offset.reset': 'earliest'
}

def consume_and_save_to_json(topic, file_path):
    c = Consumer(consumer_config)

    # Subscribe to the specified topic
    c.subscribe([topic])

    try:
        while True:
            msg = c.poll(timeout=1.0)
            if msg is None:
                continue
            if msg.error():
                if msg.error().code() == KafkaError._PARTITION_EOF:
                    # End of partition, consumer reached end of the topic
                    break
                elif msg.error():
                    raise KafkaException(msg.error())
            else:
                try:
                    # Deserialize JSON data
                    data = json.loads(msg.value().decode('utf-8'))
                    
                    # Append data to the JSON file
                    with open(file_path, 'a') as json_file:
                        json.dump(data, json_file)
                        json_file.write('\n')  # Add newline character to separate JSON objects
                    
                    print("Data appended to the file successfully.")
                    
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON: {e}")
                    continue  # Skip to the next message

    except KeyboardInterrupt:
        pass

    finally:
        # Close the consumer
        c.close()

# Specify the topics and file paths for each JSON file
topics_and_file_paths = {
    'weather': 'weather.json',
    'Departures': 'Departures.json',
    'Arrivals': 'Arrivals.json'
}

# Start consuming data from Kafka and saving it to JSON files
for topic, file_path in topics_and_file_paths.items():
    consume_and_save_to_json(topic, file_path)
```
